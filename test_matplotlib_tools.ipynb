```python
import matplotlib_tools as mpt
import matplotlib.pyplot as plt
import pandas as pd
from PIL import Image
import os
```

```python
#open images
_imgs = ['sample1.png','sample2.png','sample3.png']
folder_source = 'images'

list_images = []
for img_name in _imgs:
    img_path = os.path.join(folder_source,img_name)
    with Image.open(img_path) as img:
        img.load()
        list_images.append(img)
```

```python
#dummy dataframe
df = pd.DataFrame(
    {
        'name'    : ['Bar1', 'Bar2', 'Bar3'],
        'value'   : [50,105,75],
        'location' : [(0,1),(2,3),(4,5)],
        'image'    : list_images
    }
)
df
```

```python
ax = df.plot.bar()
```

```python
# fig, ax = plt.subplots()
plt.imshow(list_images[0])
```

```python
mpt.bar_plot_with_images(df['value'], df['image'], xtick_labels=['p1','p2','p3'], orientation='v', img_scale=0.10, x_offset=0, y_offset=-40, xlabel='X label',)
```

```python
ax.plot(df['value'], df.index)
```
